```python
# representing 3rd adhoc query
df = pd.read_csv('adhoc_request_3.csv')
ax1 = df.plot.bar(x='segment', xlabel='Segment')
ax1.set_yticks(range(0, df['product_count'].max() + 1, 2))
plt.xticks(rotation=30)
ax1.set_title('Unique product counts per Segment')
plt.grid(axis='y')
plt.show()
```

```python
# representing 4th adhoc query
df = pd.read_csv('adhoc_request_4.csv')
ax2 = df[['segment', 'count_2020', 'count_2021']].plot.bar(x='segment')
plt.xticks(rotation=30)
ax2.set_title('Segment with highest increase in unique product count from 2020 to 2021')
plt.ylim(0, 20)
plt.grid(axis='y')
plt.show()
df
```

```python
# representing 7th adhoc query
df = pd.read_csv('adhoc_request_7.csv')
df = df.set_index(['month', 'year'])
ax3 = df.plot.line(marker='o', xlabel='(month, year)')
ax3.set_title('Gross sales amount for each year in fiscal years 2020-2021')
plt.xticks(rotation=30)
plt.grid()
plt.show()
```

```python
# representing 8th adhoc query
df = pd.read_csv('adhoc_request_8.csv')
df = df.set_index('quarter')
plt.pie(df['total_sold_quantity'], labels=df.index, autopct='%.1f', explode=[0.1, 0, 0, 0], shadow=True,
             startangle=90)
plt.title('Quarters in fiscal year 2021 and their contribution')
df
```

```python
# representing 9th adhoc query
df = pd.read_csv('adhoc_request_9.csv')
df = df.set_index('channel')
plt.pie(df['gross_sales_mln'], labels=df.index, autopct='%.1f', explode=[0, 0, 0.1], shadow=True)
plt.title('Channel sales in fiscal year 2021 and their contribution')
df
```

```python
# representing 10th adhoc query
df = pd.read_csv('adhoc_request_10.csv')
df = df.set_index(['product_code', 'product'])
for title, group in df.groupby('division'):
    ax = group.plot.bar(y='total_sold')
    ax.set_title(f'Top 3 products for division {title} in units sold')
    ax.set(xlabel=None)
    plt.xticks(rotation=15)
df
```
